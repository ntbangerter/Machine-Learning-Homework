For this assignment, we'll be doing the Kaggle Titanic: Machine Learning from Disaster Competition. The objective is to predict which individuals on the Titanic survived, using the passenger's class, name, sex, age, # of siblings/spouses on the titanic, # of parents/children on the titanic, their ticket number, amount they paid in fare, their cabin number, and which port they embarked from.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import missingno as msno

import warnings
warnings.filterwarnings('ignore')

from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import f1_score, classification_report, accuracy_score
```

```python
train = pd.read_csv('train.csv')
train.head()
```

```python
test = pd.read_csv('test.csv')
test.head()
```

Everything looks good for the testing and training data we read in above. The first thing we'll do is clean the data a bit by checking for missing entries, drop columns we don't need, and change the format of some columns.

I'll check for missing data using the missingno library, created by ResidentMario (https://github.com/ResidentMario/missingno):

```python
msno.matrix(train)
plt.title("Missing Values by Category, Training Data", fontsize=24, y=1.15)
```

```python
msno.matrix(test)
plt.title("Missing Values by Category, Testing Data", fontsize=24, y=1.15)
```

It looks like age and cabin have a lot of missing values for both the testing and the training set, while the testing set has one value missing in fare and training has two values missing in embarked. For the numerical data, I'll fill the missing values with the mean from the training dataset because there doesn't seem to be a good way to infer the missing values from the data I do have. Embarked I'll fill with the most common value (S). I was tempted to drop cabin since so much of the data was missing, but I'm loathe to lose potentially relevant data when I already have so little, so I decided to fill the missing values with "U" for unknown, and change the rest of the values with the first letter from the cabin, which I found out by looking up maps of Titanic online refers to the deck the cabin is located on.

```python
#cleaning cabin data

train.loc[train['Cabin'].isnull(),'Cabin'] = "U"

for i in range(0, len(train)):
    train['Cabin'][i] = train['Cabin'][i][0]
    
test.loc[test['Cabin'].isnull(),'Cabin'] = "U"

for i in range(0, len(test)):
    test['Cabin'][i] = test['Cabin'][i][0]

train.head()
```

```python
#filling missing values in age, embarked, and fare
train.loc[train['Age'].isnull(),'Age'] = train['Age'].mean()
train.loc[train['Embarked'].isnull(),'Embarked'] = 'S'

test.loc[test['Age'].isnull(),'Age'] = train['Age'].mean()
test.loc[test['Fare'].isnull(),'Fare'] = train['Fare'].mean()
```

Now that I've filled the missing values, I'll drop the columns I don't think are relevant. Passenger ID certainly isn't useful. Name could be useful but would take a significant amount of effort to extract data from, so I'm going to drop it. Ticket number, as far as I can tell, doesn't seem to have much of a pattern and shouldn't make a difference on survival rates, so I'll drop it as well.

```python
#set the target metric
target = train['Survived']

#create the training and testing data set with only the relevant variables.
train_data = train.drop(['PassengerId','Survived','Name','Ticket'], axis=1)
test_data = test.drop(['PassengerId','Name','Ticket'], axis=1)

train_data.head()
```

```python
test_data.head()
```

I tried a number of bucketing methods for age and fare, but none gave me substantially different results from leaving them as continuous values for my classification models, so I ended up just leaving them in their continuous form. I made dummy variables of the categorical data for running my regressions:

```python
train_data = pd.get_dummies(train_data)
test_data = pd.get_dummies(test_data)
```

To make sure the columns are identical in my training and testing data set, I checked for any column that was in one but not the other:

```python
train_data.columns.difference(test_data.columns)
```

```python
test_data.columns.difference(train_data.columns)
```

The training data set had an extra column from a dummy variable for Cabin, which I can safely drop because the data won't be useful for predicting the testing data since the column isn't present in the testing data set.

```python
train_data = train_data.drop('Cabin_T', axis=1)
train_data.columns.difference(test_data.columns)
```

I'm going to create a correlation matrix for my data, to see if I can identify which columns will be most useful in predicting survival. At the end, I'll use an analysis of a random forest's binary trees to see how well I was able to predict the useful variables.

```python
correlation_matrix = pd.DataFrame({
    "Survived": target,
    "Pclass": test_data["Pclass"],
    "Age": test_data["Age"],
    "SibSp": test_data["SibSp"],
    "Parch": test_data["Parch"],
    "Fare": test_data["Fare"],
    "Sex_female": test_data["Sex_female"],
    "Cabin_U": test_data["Cabin_U"],
    "Embarked_C": test_data["Embarked_C"],
    "Embarked_S": test_data["Embarked_S"],
    "Embarked_Q": test_data["Embarked_Q"],
})

plt.figure(figsize=(14,12))
plt.title('Correlation Matrix', y=1.05, size=16)
sns.heatmap(correlation_matrix.astype(float).corr(),linewidths=0.1,vmax=1.0, 
            square=True, linecolor='white', annot=True)
```

From the heatmap, it looks like Cabin_U, Embarked_Q, Sex, SibSp, and Pclass are the most influential columns. We'll check this at the end of the project.

Now that the data has been prepared for modeling, I can create my train/test split for cross validation and start trying a variety of models to see which performs best. I'll try logistic regression, k-nearest neighbors, SVC, random forest, gradient boosting, and AdaBoost.

```python
X_train, X_test, y_train, y_test = train_test_split(train_data, target, test_size=0.33, random_state=42)
```

### Logistic Regression

```python
logistic_regression = GridSearchCV(LogisticRegression(), {'penalty': ['l1','l2'], 
                                                          "C": [0.025,0.05,0.075,0.25,0.5]})
logistic_regression.fit(X_train, y_train)

train_predictions = logistic_regression.predict(X_train)
test_predictions = logistic_regression.predict(X_test)

print("Best Params: {}".format(logistic_regression.best_params_))
print("Train F1: {}".format(f1_score(y_train, train_predictions)))
print("Test Classification Report:")
print(classification_report(y_test, test_predictions))
print("Train Accuracy: {}\tTest accuracy: {}".format(accuracy_score(y_train, train_predictions),
                                                     accuracy_score(y_test, test_predictions)))
```

### K-Nearest Neighbors

```python
k_neighbors = GridSearchCV(KNeighborsClassifier(), {'n_neighbors': [1,3,5,7,12,15,20], 
                                                    "weights": ["uniform","distance"], 
                                                    "p": [1,2,3]})
k_neighbors.fit(X_train, y_train)

train_predictions = k_neighbors.predict(X_train)
test_predictions = k_neighbors.predict(X_test)

print("Best Params: {}".format(k_neighbors.best_params_))
print("Train F1: {}".format(f1_score(y_train, train_predictions)))
print("Test Classification Report:")
print(classification_report(y_test, test_predictions))
print("Train Accuracy: {}\tTest accuracy: {}".format(accuracy_score(y_train, train_predictions),
                                                     accuracy_score(y_test, test_predictions)))
```

### SVC

```python
svc = GridSearchCV(SVC(), {'kernel': ["linear"], 
                           "C": [0.025]})
svc.fit(X_train, y_train)

train_predictions = svc.predict(X_train)
test_predictions = svc.predict(X_test)

print("Best Params: {}".format(svc.best_params_))
print("Train F1: {}".format(f1_score(y_train, train_predictions)))
print("Test Classification Report:")
print(classification_report(y_test, test_predictions))
print("Train Accuracy: {}\tTest accuracy: {}".format(accuracy_score(y_train, train_predictions),
                                                     accuracy_score(y_test, test_predictions)))
```

### Random Forest

```python
random_forest = GridSearchCV(RandomForestClassifier(), {'n_estimators': [10,20,50,100,250,500], 
                                                        "max_depth": [2,4,6,10], 
                                                        "min_samples_leaf": [2,5]})
random_forest.fit(X_train, y_train)

train_predictions = random_forest.predict(X_train)
test_predictions = random_forest.predict(X_test)

print("Best Params: {}".format(random_forest.best_params_))
print("Train F1: {}".format(f1_score(y_train, train_predictions)))
print("Test Classification Report:")
print(classification_report(y_test, test_predictions))
print("Train Accuracy: {}\tTest accuracy: {}".format(accuracy_score(y_train, train_predictions),
                                                     accuracy_score(y_test, test_predictions)))
```

### Gradient Boosting

```python
gradient_boosting = GridSearchCV(GradientBoostingClassifier(), {'n_estimators': [10,20,50,100,250,500]})
gradient_boosting.fit(X_train, y_train)

train_predictions = gradient_boosting.predict(X_train)
test_predictions = gradient_boosting.predict(X_test)

print("Best Params: {}".format(gradient_boosting.best_params_))
print("Train F1: {}".format(f1_score(y_train, train_predictions)))
print("Test Classification Report:")
print(classification_report(y_test, test_predictions))
print("Train Accuracy: {}\tTest accuracy: {}".format(accuracy_score(y_train, train_predictions),
                                                     accuracy_score(y_test, test_predictions)))
```

### AdaBoost

```python
ada_boost = GridSearchCV(AdaBoostClassifier(), {'n_estimators': [10,20,50,100,250,500], 
                                                "learning_rate": [0.01,0.03,0.1,0.3,1]})
ada_boost.fit(X_train, y_train)

train_predictions = ada_boost.predict(X_train)
test_predictions = ada_boost.predict(X_test)

print("Best Params: {}".format(ada_boost.best_params_))
print("Train F1: {}".format(f1_score(y_train, train_predictions)))
print("Test Classification Report:")
print(classification_report(y_test, test_predictions))
print("Train Accuracy: {}\tTest accuracy: {}".format(accuracy_score(y_train, train_predictions),
                                                     accuracy_score(y_test, test_predictions)))
```

Most of the models were pretty comparable in terms of precision, recall, and f1-score. Most were in the range of 0.79-0.80 for all three metrics, with the exception of K-Nearest Neighbors, which scored significantly lower, and Gradient Boosting, which scored slightly higher. For this analysis, I'm primarily interested in accuracy since that's what Kaggle will be using as its evaluation metric, but for other analyses it would be useful to choose an appropriate metric to suit the problem at hand. 

## Submission to Kaggle

Logistic regression, random forest, and gradient boosting all gave me pretty high test accuracy levels relative to the rest, so I'm going to train a model for each of them using the full training set with their best cross-validated parameters and use each to predict the values for my submission to Kaggle.

```python
logistic_regression = LogisticRegression(C=0.25, penalty='l2')
logistic_regression.fit(train_data, target)

random_forest = RandomForestClassifier(max_depth=10, min_samples_leaf=5, n_estimators=50)
random_forest.fit(train_data, target)

gradient_boosting = GradientBoostingClassifier(n_estimators=100)
gradient_boosting.fit(train_data, target)
```

```python
rf_predictions = random_forest.predict(test_data)

rf_submission = pd.DataFrame({
    "PassengerId": test['PassengerId'],
    "Survived": rf_predictions
})
rf_submission.to_csv('rf_submission.csv', index=False)
```

```python
lr_predictions = logistic_regression.predict(test_data)

lr_submission = pd.DataFrame({
    "PassengerId": test['PassengerId'],
    "Survived": lr_predictions
})
lr_submission.to_csv('lr_submission.csv', index=False)
```

```python
gb_predictions = gradient_boosting.predict(test_data)

gb_submission = pd.DataFrame({
    "PassengerId": test['PassengerId'],
    "Survived": gb_predictions
})
gb_submission.to_csv('gb_submission.csv', index=False)
```

Both the gradient boosting and random forest's predictions scored 0.77033 on Kaggle, with logistic regression coming in slightly lower at 0.76555. Given that the models disagreed on quite a few of the results, I decided to try a compounded prediction using the three of them. Since each of the models was essentially as accurate as the next, I weighted them all equally and created a new submission. This submission ended up tying with gradient boosting and random forest, which was a little odd, so I didn't manage to get any higher using that. My best rank ended up being 6234 on the Kaggle leaderboards.

```python
overall_predictions = (rf_predictions + lr_predictions + gb_predictions) / 3

for i in range(0, len(overall_predictions)):
    if overall_predictions[i] > 0.5:
        overall_predictions[i] = 1
    else:
        overall_predictions[i] = 0
        
overall_submission = pd.DataFrame({
    "PassengerId": test['PassengerId'],
    "Survived": overall_predictions
})
overall_submission.to_csv('overall_submission.csv', index=False)
```

As a little post-analysis, I'll use the Random Forest to identify which features it ended up giving the most weight to when determining outcome:

```python
feature_imp = sorted(list(zip(list(train_data), random_forest.feature_importances_)), key=lambda x: x[1], reverse=True)
```

```python
plt.subplots(figsize=(12,6))
plt.ylabel("Relative Importance")
plt.title("Most Important Variables, Random Forest", size=16, y=1.05)
pd.Series([x[1] for x in feature_imp], index=[x[0] for x in feature_imp]).plot(kind='bar')
```

While gender was the most important feature, I actually did pretty bad at predicting which features would be useful from the correlation matrix. Age had a low correlation, though that's primarily because age didn't have a linear relationship with survival. Pclass and Cabin_U I guessed would be pretty high, but I didn't expect Fare to be so important, and definitely not more important than Pclass. A lot of the other features, like Embarked and Cabins A-G had almost no relative importance, which I found a bit surprising given that I expected the Cabins to correlate some with location on the ship, which I imagined would have an impact on making it off alive or not. All told, I was rather surprised by which features ended up being most important in the regression.
